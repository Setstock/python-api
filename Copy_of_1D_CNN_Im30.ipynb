```python
import numpy as np
import pandas as pd

# Iterate through the columns from index 3 to 513
for column in processed_df.columns[0:5]:
    # Calculate quartiles for the current column
    quartiles = processed_df[column].quantile([0.25, 0.5, 0.75])

    # Iterate through the rows in the current column
    for index, value in enumerate(processed_df[column]):
        if 0.0 <= value <= quartiles[0.25]:
            processed_df.at[index, column] = "A"
        elif quartiles[0.25] < value <= quartiles[0.5]:
            processed_df.at[index, column] = "B"
        elif quartiles[0.5] < value <= quartiles[0.75]:
            processed_df.at[index, column] = "C"
        else:
            processed_df.at[index, column] = "D"
```

```python
# Concatenate the columns into a new 'gencode' column
processed_df['gencode'] = processed_df.apply(lambda row: ''.join(row[['ราคาสูงสุด', 'ราคาต่ำสุด', 'volume', 'A/B', 'P/R']]), axis=1)

# Display the DataFrame
print(processed_df)
```

```python
from keras.utils import to_categorical

# Assuming df is your DataFrame with the data
text_data = processed_df['gencode'].tolist()
labels = processed_df['label'].tolist()

# Initialize the tokenizer
tokenizer = Tokenizer(char_level=True)

# Fit the tokenizer on the text data
tokenizer.fit_on_texts(text_data)

# Convert text to sequences
sequences = tokenizer.texts_to_sequences(text_data)

# Get the word index mapping
word_index = tokenizer.word_index


# Pad sequences to make them uniform length (optional)
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')#xissequence

# Display the results
print("Word Index:", word_index)
print("Sequences:", sequences)
print("Padded Sequences:", padded_sequences)

# One-hot encode the labels
one_hot_labels = to_categorical(labels) #yisonehot

# Display the one-hot encoded labels
print("One-Hot Encoded Labels:")
print(one_hot_labels)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

import numpy as np

# Example values for max_length and num_features
max_length = 64  # Length of the input sequence
num_features = 100  # Number of features per time step
n_outputs = 1  # Number of output classes

# Generate random input data for testing
# For example, let's say we have 1000 samples
num_samples = 1000
X_train = np.random.rand(num_samples, max_length, num_features)
y_train = np.random.randint(0, 2, size=(num_samples, n_outputs))  # Assuming binary classification

# Define the model
verbose, epochs, batch_size = 0, 10, 32
model = Sequential()

model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(max_length, num_features), padding='same'))
print("After Conv1D-1:", model.output_shape)
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'))
print("After Conv1D-2:", model.output_shape)
model.add(Dropout(0.5))
print("After Dropout:", model.output_shape)
model.add(MaxPooling1D(pool_size=2, padding='same'))
print("After MaxPooling1D:", model.output_shape)
model.add(Flatten())
print("After Flatten:", model.output_shape)
model.add(Dense(100, activation='relu'))
print("After Dense-1:", model.output_shape)
model.add(Dense(n_outputs, activation='softmax'))
print("After Dense-2:", model.output_shape)

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, batch_size=batch_size, verbose=verbose, epochs=epochs)

# Test the model with input data
# model.predict(input_data)
```

```python
# Example values for max_length and num_features
max_length = 64  # Length of the input sequence
num_features = 100  # Number of features per time step
n_outputs = 2  # Number of output classes (for binary classification)

# Generate random input data for testing
num_samples = 1000
padded_sequences = np.random.rand(num_samples, max_length, num_features)
labels = np.random.randint(0, n_outputs, size=(num_samples,))
one_hot_labels = to_categorical(labels, num_classes=n_outputs)

# Training and test data
X_train = padded_sequences
y_train = one_hot_labels

X_test = np.random.rand(200, max_length, num_features)
y_test = np.random.randint(0, n_outputs, size=(200,))
y_test_one_hot = to_categorical(y_test, num_classes=n_outputs)

# Define the model
verbose, epochs, batch_size = 0, 10, 32
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(max_length, num_features), padding='same'))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'))
model.add(Dropout(0.5))
model.add(MaxPooling1D(pool_size=2, padding='same'))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(n_outputs, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(X_train, y_train, batch_size=batch_size, verbose=verbose, epochs=epochs)

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test_one_hot)
print('Test Loss: {}'.format(test_loss))
print('Test Accuracy: {}'.format(test_acc))

# Make predictions on the test set
y_pred = model.predict(X_test)

# Convert predicted probabilities to binary predictions
y_pred_classes = np.argmax(y_pred, axis=1)

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred_classes, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred_classes, average='weighted', zero_division=0)
F1 = f1_score(y_test, y_pred_classes, average='weighted', zero_division=0)

print('Precision: {}'.format(precision))
print('Recall: {}'.format(recall))
print('F1: {}'.format(F1))

# Optionally, print a detailed classification report
report = classification_report(y_test, y_pred_classes)
print(report)
```
